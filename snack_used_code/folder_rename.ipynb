```python
# snack_case의 폴더이름을 camel_case의 폴더이름으로 바꾸는 코드
import os

# 스네이크 케이스를 카멜 케이스로 변환하는 함수
def snake_to_camel(snake_str):
    components = snake_str.split('_')
    return ''.join(x.title() for x in components)

# 폴더가 있는 경로
folder_path = "D:/classification_data/train/Quant_train_labels" # 실제 폴더 경로를 적어주세요.

# 현재 스네이크 케이스 폴더 이름을 카멜 케이스로 변경
for snake_case_name in os.listdir(folder_path):
    # 스네이크 케이스 폴더 이름을 카멜 케이스로 변환
    camel_case_name = snake_to_camel(snake_case_name)

    old_folder_path = os.path.join(folder_path, snake_case_name)
    new_folder_path = os.path.join(folder_path, camel_case_name)

    # 폴더 이름을 변경
    if os.path.exists(old_folder_path):
        os.rename(old_folder_path, new_folder_path)
        print(f"Renamed {snake_case_name} to {camel_case_name}")
    else:
        print(f"{snake_case_name} does not exist")
```
