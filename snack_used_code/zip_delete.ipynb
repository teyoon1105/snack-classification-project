```python
# 폴더 안에 존재하는 모든 zip파일 삭제하기
import os
import shutil

def delete_zip_files(directory_path):
    """
    지정된 디렉토리 안의 모든 zip 파일을 삭제합니다.

    Args:
        directory_path: 삭제할 zip 파일이 있는 디렉토리 경로
    """

    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        if os.path.isfile(file_path) and filename.endswith(".zip"):
            try:
                os.remove(file_path)
                print(f"Deleted: {filename}")
            except Exception as e:
                print(f"Error deleting {filename}: {e}")

train_path = "D:/AI_HUB/TRAIN/[라벨]양추정_TRAIN/txt"

food_list = os.listdir(train_path)
for food in food_list:
    food_path = os.path.join(train_path, food)
    quant_list = os.listdir(food_path)
    for quant in quant_list:
        quant_path = os.path.join(food_path, quant)
        delete_zip_files(quant_path)
```
