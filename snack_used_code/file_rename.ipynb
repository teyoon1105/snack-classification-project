```python
# 한글 이름의 파일을 영어 이름(camel_case)로 바꾸기

import os

# 상위경로, train과 valid를 분리하는 첫번째 폴더 경로
# train_image_path = "D:/classification_data/train/Quant_train_labels"
train_image_path = "D:/classification_data/train/Quant_train_images"

# 해당 폴더에 들어가면 영문으로 된 음식 메뉴 폴더들이 83개 존재
# 폴더 내 메뉴 리스트를 가져옴
menu_list = os.listdir(train_image_path)

# 각 메뉴 이름을 저장한 리스트에서 for문을 통해 하나씩 가져오고
for menu in menu_list:
    # 메뉴 폴더에 접근할 수 있는 경로를 만듦
    menu_path = os.path.join(train_image_path, menu)
    # 해당 메뉴 폴더 안에는 양추정을 나누는 Q1 ~ Q5 폴더가 존재
    # 해당 폴더 이름을 리스트에 저장
    q_list = os.listdir(menu_path)

    # Q1 ~ Q5 까지 for문을 통해 가져옴
    for q in q_list:
        # 각 Q폴더에 접근할 수 있는 경로를 만듦
        q_path = os.path.join(menu_path, q)
        # 해당 폴더 안에는 한글로 되어있는 이미지 파일들이 존재
        # 해당 파일의 이름들을 전부 리스트에 저장
        food_list = os.listdir(q_path)

        # 파일 이름을 바꾸는 부분

        # 한글로 된 음식 파일의 음식 이름을 for문을 통해 가져옴
        for food in food_list:
            # 해당 파일에 접근할 경로를 만듦
            food_path = os.path.join(q_path, food)

            # 파일 이름을 '_' 기준으로 split
            file_parts = food.split('_')
            # file_parts = ['side', '국밥류', '배달용기그릇', '갈비탕', 'Q1', '00001']
            # 여기서 버릴 껀 1번, 2번, 3번을 버리고, 버린 부분에 위에 메뉴 이름을 넣길 원함
            
            
            # 새 파일 이름 만들기: 'side', 'menu', 'Q1', '00001' 유지
            if len(file_parts) >= 6:
                # 확장자를 포함한 원본 파일 이름 그대로 사용
                new_file_name = f"{file_parts[0]}_{menu}_{file_parts[4]}_{file_parts[5]}"
                # 그럼 new_file_name = side_Galbitang_Q1_00001
                # 이렇게 변함
                
                new_file_path = os.path.join(q_path, new_file_name)
                # 변할 수 있게 경로를 만들어주고

                # 파일 이름 변경
                os.rename(food_path, new_file_path)
                print(f"Renamed {food} to {new_file_name}")
            
        
```

```python
# 영어 이름(snack case)의 파일을 영어 이름(camel case)로 바꾸기
# 실패한 케이스
# 실패한 이휴
# 아래 코드는 menu리스트가 결국 언더바에 대해 split을 해도 하나의 단어로 나온다는 착각에
# 아무 생각없이 실행을 했다가 split에 의해 잘못 나뉘어져서 파일들 끼리 구분이 안되어 하나로 덮어 씌워진 케이스
# ex : side_Tomato_Sauce_Spaghetti_Q1_00001 와 side_Albap_Q1_00001과 비교
# split으로 나누면 [side, Tomato, Sauce, Spaghetti, Q1, 00001] vs [side, Albap, Q1, 00001]
# 때문에 Albap에 맞게 코드를 짰기 때문에
# 전자의 경우는 Q1과 00001이 파일 이름에 들어가는 대신 다시 Sauce와 Spaghetti가 나오게 됨
import os

train_image_path = "D:/classification_data/valid/Quant_valid_labels"

# 폴더 내 메뉴 리스트를 가져옴
menu_list = os.listdir(train_image_path)

for menu in menu_list:
    menu_path = os.path.join(train_image_path, menu)
    q_list = os.listdir(menu_path)

    for q in q_list:
        q_path = os.path.join(menu_path, q)
        food_list = os.listdir(q_path)

        # 파일 이름을 바꾸는 부분
        for food in food_list:
            food_path = os.path.join(q_path, food)

            # 파일 이름을 '_' 기준으로 split
            file_parts = food.split('_')
            
            # 새 파일 이름 만들기: 'side', 'menu', 'Q1', '00001' 유지
            if len(file_parts) >= 4:
                # 확장자를 포함한 원본 파일 이름 그대로 사용
                new_file_name = f"{file_parts[0]}_{menu}_{file_parts[2]}_{file_parts[3]}"
                
                new_file_path = os.path.join(q_path, new_file_name)

                # 파일 이름 변경
                os.rename(food_path, new_file_path)
                print(f"Renamed {food} to {new_file_name}")
```

```python
# 갈치조림의 파일 이름이 특이해서 따로 짜준 드드
import os

# 갈치조림 폴더를 경로에 넣는다
menu_path = "D:/classification_data_resize/valid/Quant_valid_labels/Galchijorim"
menu = 'Galchijorim'
q_list = os.listdir(menu_path)

# Q1 ~ Q5 까지 for문을 통해 가져옴
for q in q_list:
    # 각 Q폴더에 접근할 수 있는 경로를 만듦
    q_path = os.path.join(menu_path, q)
    # 해당 폴더 안에는 한글로 되어있는 이미지 파일들이 존재
    # 해당 파일의 이름들을 전부 리스트에 저장
    food_list = os.listdir(q_path)

    # 한글로 된 음식 파일의 음식 이름을 for문을 통해 가져옴
    for food in food_list:
        # 해당 파일에 접근할 경로를 만듦
        food_path = os.path.join(q_path, food)

        # 파일 이름을 '_' 기준으로 split
        file_parts = food.split('_')
        
        # Q부분과 번호를 올바른 형식으로 변환하는 함수
        def fix_q_format(q_string):
            if q_string.startswith("Q") and len(q_string) > 2:
                # Q와 나머지 숫자 부분 분리
                q_number = q_string[1]
                number_part = q_string[2:].split('.')[0]  # 확장자 제거
                # 숫자 부분을 5자리로 맞춤
                formatted_number = f"{int(number_part):05d}"
                # Q1_00001 형식으로 반환
                return f"Q{q_number}_{formatted_number}"
            return q_string  # 형식에 맞지 않으면 그대로 반환
        
        # 파일 이름을 제대로 바꾸기 위해, Q부분을 변환
        if len(file_parts) >= 5:
            # Q부분 수정
            fixed_q = fix_q_format(file_parts[4])
            # 새 파일 이름 만들기
            new_file_name = f"{file_parts[0]}_{menu}_{fixed_q}"
            # 파일 확장자 추가
            extension = os.path.splitext(food)[1]
            new_file_name += extension
            # 새로운 파일 경로 설정
            new_file_path = os.path.join(q_path, new_file_name)

            # 파일 이름 변경
            os.rename(food_path, new_file_path)
            print(f"Renamed {food} to {new_file_name}")
        
    


```

```python
# Ssambap이라고 잘못 줬기 때문에
# 이를 Ssalbap으로 고치는 드드
import os

# 최상위 디렉토리 경로
base_path = "D:/classification_data_resize"

# 변경할 문자열 설정
old_str = "Ssambap"
new_str = "Ssalbap"

# 디렉토리를 순회하면서 파일 및 폴더 이름 변경
# topdown=False --> 가장 하위 디렉토리부터 순회 > 상위 디렉토리 이름을 바꿀 때 문제가 생기지 않음
for root, dirs, files in os.walk(base_path, topdown=False):
    # 파일 이름 변경
    for file_name in files:
        if old_str in file_name:
            # 기존 파일 경로와 새로운 파일 경로 생성
            old_file_path = os.path.join(root, file_name)
            new_file_name = file_name.replace(old_str, new_str)
            new_file_path = os.path.join(root, new_file_name)
            # 파일 이름 변경
            os.rename(old_file_path, new_file_path)
            print(f"Renamed file: {old_file_path} -> {new_file_path}")

    # 폴더 이름 변경
    for dir_name in dirs:
        if old_str in dir_name:
            # 기존 폴더 경로와 새로운 폴더 경로 생성
            old_dir_path = os.path.join(root, dir_name)
            new_dir_name = dir_name.replace(old_str, new_str)
            new_dir_path = os.path.join(root, new_dir_name)
            # 폴더 이름 변경
            os.rename(old_dir_path, new_dir_path)
            print(f"Renamed directory: {old_dir_path} -> {new_dir_path}")

```
