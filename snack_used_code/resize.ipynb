```python
import cv2
import os

# resize 할 너비와 높이 설정
width = 640
height = 640

# train_images 폴더 경로
train_path = "D:/classification_data_resize/valid/Quant_valid_images"

# 폴더 속 83개 클래스의 이름을 리스트로 저장
menu_list = os.listdir(train_path)

# 각 클래스 이름을 하나씩 for문을 통해 가져옴
for menu in menu_list:
    # 각 폴더에 접근할 경로 만들기
    menu_path = os.path.join(train_path, menu)
    q_list = os.listdir(menu_path)
    for q in q_list:
        q_path = os.path.join(menu_path, q)
        food_list = os.listdir(q_path)
        for food in food_list:
            imgpath = os.path.join(q_path, food)
            # 이미지 읽기
            img = cv2.imread(imgpath)
            
            # 이미지가 제대로 읽혔는지 확인
            if img is None:
                print(f"이미지를 읽을 수 없습니다. 경로를 확인하세요: {imgpath}")
                continue
            
            # 이미지의 현재 크기를 확인
            current_height, current_width = img.shape[:2]
            
            # 이미 크기가 640x640인 경우 resize 생략
            if current_width == width and current_height == height:
                print(f"이미지는 이미 {width}x{height} 크기입니다: {imgpath}")
                continue
            
            # 이미지 크기 조정
            resized_image = cv2.resize(img, (width, height), interpolation=cv2.INTER_LANCZOS4)
            
            # 이미지 저장 (덮어쓰기)
            cv2.imwrite(imgpath, resized_image)
            
            print(f"이미지가 성공적으로 크기 조정 및 저장되었습니다: {imgpath}")
```

```python
# 형과 내가 찍은 사진(etc추가된) 파일 resize하는 코드

import os
import cv2

# basepath 설정
# basepath = "D:/validation/07014001_valid"  # 실제 데이터 경로로 변경해주세요.
basepath = "D:/food_detection/train_16/04011005"


# resize 목표 크기
target_size = (640, 640)

# sam2 resize
# target_size = (1024, 1024)

# basepath 아래의 모든 파일 순회
for filename in os.listdir(basepath):
    # 이미지 파일 경로
    image_path = os.path.join(basepath, filename)
    
    # 이미지 불러오기
    image = cv2.imread(image_path)

    # resize 수행
    resized_image = cv2.resize(image, target_size)

    # resize된 이미지를 원본 이미지 파일 경로에 덮어쓰기
    cv2.imwrite(image_path, resized_image)

    print(f"{filename} 이미지 resize 완료.")


print("모든 이미지 resize 완료.")
```
