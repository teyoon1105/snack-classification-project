```python
import os
import shutil

# 원본 train 디렉토리 경로
train_dir = "D:/classification_data_resize/valid/Quant_valid_labels"


# train 디렉토리의 각 메뉴 폴더에 접근
for menu_folder in os.listdir(train_dir):
    menu_path = os.path.join(train_dir, menu_folder)
    
    if os.path.isdir(menu_path):
        # Q1 ~ Q5 폴더 탐색
        for q_folder in os.listdir(menu_path):
            q_folder_path = os.path.join(menu_path, q_folder)
            
            if os.path.isdir(q_folder_path):
                # 각 Q 폴더의 이미지 파일을 대상 디렉토리로 옮김
                for img_file in os.listdir(q_folder_path):
                    img_path = os.path.join(q_folder_path, img_file)
                    
                    if os.path.isfile(img_path):
                        # 파일 이름에 Q 폴더 정보 포함시키기 (예: Q1_image1.jpg)
                        target_path = os.path.join(menu_path, img_file)
                        shutil.move(img_path, target_path)
                    
            
    
            

print("이미지 이동 완료!")
```

```python
import os
import shutil

# 원본 train 디렉토리 경로
train_dir = "D:/classification_data_resize/valid/Quant_valid_labels"

# train 디렉토리의 각 메뉴 폴더에 접근
for menu_folder in os.listdir(train_dir):
    menu_path = os.path.join(train_dir, menu_folder)
    
    if os.path.isdir(menu_path):
        # Q1 ~ Q5 폴더 탐색 및 삭제
        for q_folder in os.listdir(menu_path):
            q_folder_path = os.path.join(menu_path, q_folder)
            
            if os.path.isdir(q_folder_path):
                # Q 폴더 삭제
                shutil.rmtree(q_folder_path)
                print(f"{q_folder} 폴더 삭제 완료")


print("모든 Q1~Q5 폴더 삭제 완료!")
```
